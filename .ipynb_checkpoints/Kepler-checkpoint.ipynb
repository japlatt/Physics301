```python
import kplr
import numpy as np
import astropy
from pylab import *
client = kplr.API()

%matplotlib inline
```

```python
planet = client.planet("Kepler-68b")
print(planet.kepoi_name)
```

```python
lc = planet.get_light_curves()
```

```python
def extract_lc(lc):
    fig = Figure(figsize = (7,5))
    time_arr = []
    flux_arr = []
    std_flux = []
    for i in xrange(len(lc)):
        time = lc[i].open()[1].data["TIME"]
        flux = lc[i].open()[1].data["PDCSAP_FLUX"]
        
        time_arr.append(time)
        flux_arr.append(flux)
        std_flux.append(np.nanstd(flux))
        
    return array(time_arr), array(flux_arr), array(std_flux)
    
```

```python
time, flux, std_flux = extract_lc(lc)
```

```python
inds = std_flux < 110
timeF = time[inds]
fluxF = flux[inds]

flux_tot = flux[0]
time_tot = time[0]
for i in xrange(1,len(flux)):
    flux_tot = concatenate((flux_tot, flux[i]/np.nanmedian(flux[i])))
    time_tot = concatenate((time_tot, time[i]))
```

```python
period1 = 5.398763
period2 = 9.605065
for i in xrange(len(timeF)):
    plot((1.8+timeF[i])%period1, fluxF[i]/np.nanmedian(fluxF[i]),'k,')
```

```python
def correlate(flux):
    coor = np.zeros(len(flux)/2)
    mean = np.nanmean(flux)
    standard_dev = np.nanstd(flux)
    for delta in xrange(len(flux)/2):
        I1 = copy(flux)
        I2 = array(list(flux[delta:]) + list(flux[0:delta]))

        inds1 = (I1 >= mean - 5*standard_dev) & (I1 <=  mean + 5*standard_dev)
        inds2 = (I2 >= mean - 5*standard_dev) & (I2 <=  mean + 5*standard_dev)
        inds = inds1 & inds2
        N = sum(inds)
        coor[delta] = sum((I1[inds]*I2[inds]))/N
    return coor
```

```python
coors = correlate(flux_tot)
```

```python
plot(coors, '.')
plt.axvline(period2*0.5)
```
